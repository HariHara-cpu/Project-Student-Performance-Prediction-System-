```python
"""
COMPLETE EDA AND HYPOTHESIS TESTING PROJECT
Student Performance Dataset Analysis
Following 7 Stages of Data Science Process
"""

# ============================================================================
# STAGE 1: PROBLEM DEFINITION
# ============================================================================
"""
OBJECTIVE: Analyze factors affecting student performance in Portuguese language course
RESEARCH QUESTIONS:
1. What factors significantly impact student grades?
2. Is there a relationship between family background and academic performance?
3. Does study time correlate with final grades?
4. Are there gender-based differences in performance?
5. How does alcohol consumption affect grades?
"""

# ============================================================================
# STAGE 2: DATA COLLECTION
# ============================================================================

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import normaltest, ttest_ind, chi2_contingency, f_oneway, pearsonr
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load the dataset
print("="*80)
print("STAGE 2: DATA COLLECTION")
print("="*80)

df = pd.read_csv('student-por.csv', sep=';')
print(f"\n✓ Dataset loaded successfully!")
print(f"Dataset shape: {df.shape}")
print(f"Number of students: {df.shape[0]}")
print(f"Number of features: {df.shape[1]}")

# ============================================================================
# STAGE 3: DATA PREPARATION & CLEANING
# ============================================================================
print("\n" + "="*80)
print("STAGE 3: DATA PREPARATION & CLEANING")
print("="*80)

# 3.1 Initial Data Inspection
print("\n--- 3.1 Dataset Overview ---")
print(df.head())

print("\n--- Data Types ---")
print(df.dtypes)

print("\n--- Dataset Information ---")
print(df.info())

# 3.2 Missing Values Analysis
print("\n--- 3.2 Missing Values Check ---")
missing_values = df.isnull().sum()
print(f"Total missing values: {missing_values.sum()}")
if missing_values.sum() > 0:
    print(missing_values[missing_values > 0])
else:
    print("✓ No missing values found!")

# 3.3 Duplicate Records
print("\n--- 3.3 Duplicate Records Check ---")
duplicates = df.duplicated().sum()
print(f"Number of duplicate records: {duplicates}")
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"✓ Removed {duplicates} duplicate records")

# 3.4 Statistical Summary
print("\n--- 3.4 Statistical Summary ---")
print(df.describe())

# 3.5 Create derived features
print("\n--- 3.5 Feature Engineering (Preliminary) ---")
# Average grade
df['avg_grade'] = (df['G1'] + df['G2'] + df['G3']) / 3

# Total alcohol consumption
df['total_alcohol'] = df['Dalc'] + df['Walc']

# Study efficiency (grade per study time)
df['study_efficiency'] = df['G3'] / (df['studytime'] + 1)  # +1 to avoid division by zero

# Family education level (average of mother and father education)
df['family_education'] = (df['Medu'] + df['Fedu']) / 2

# Performance categories
df['performance_category'] = pd.cut(df['G3'], 
                                     bins=[0, 9, 14, 20], 
                                     labels=['Poor', 'Average', 'Good'])

print("✓ Created new features: avg_grade, total_alcohol, study_efficiency, family_education, performance_category")

# ============================================================================
# STAGE 4: EXPLORATORY DATA ANALYSIS (EDA)
# ============================================================================
print("\n" + "="*80)
print("STAGE 4: EXPLORATORY DATA ANALYSIS")
print("="*80)

# 4.1 Univariate Analysis - Numerical Variables
print("\n--- 4.1 Univariate Analysis: Numerical Variables ---")

numerical_cols = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 
                  'failures', 'famrel', 'freetime', 'goout', 'Dalc', 
                  'Walc', 'health', 'absences', 'G1', 'G2', 'G3']

fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    axes[idx].hist(df[col], bins=20, edgecolor='black', alpha=0.7, color='skyblue')
    axes[idx].set_title(f'Distribution of {col}', fontsize=10, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('univariate_numerical.png', dpi=300, bbox_inches='tight')
print("✓ Saved: univariate_numerical.png")
plt.close()

# 4.2 Univariate Analysis - Categorical Variables
print("\n--- 4.2 Univariate Analysis: Categorical Variables ---")

categorical_cols = ['school', 'sex', 'address', 'famsize', 'Pstatus', 
                    'schoolsup', 'famsup', 'paid', 'activities', 
                    'nursery', 'higher', 'internet', 'romantic']

fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for idx, col in enumerate(categorical_cols):
    value_counts = df[col].value_counts()
    axes[idx].bar(value_counts.index, value_counts.values, color='coral', edgecolor='black')
    axes[idx].set_title(f'Distribution of {col}', fontsize=10, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Count')
    axes[idx].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('univariate_categorical.png', dpi=300, bbox_inches='tight')
print("✓ Saved: univariate_categorical.png")
plt.close()

# 4.3 Target Variable Analysis (G3 - Final Grade)
print("\n--- 4.3 Target Variable Analysis: Final Grade (G3) ---")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram
axes[0].hist(df['G3'], bins=20, edgecolor='black', alpha=0.7, color='green')
axes[0].set_title('Distribution of Final Grades (G3)', fontweight='bold')
axes[0].set_xlabel('Grade')
axes[0].set_ylabel('Frequency')
axes[0].axvline(df['G3'].mean(), color='red', linestyle='--', label=f'Mean: {df["G3"].mean():.2f}')
axes[0].axvline(df['G3'].median(), color='blue', linestyle='--', label=f'Median: {df["G3"].median():.2f}')
axes[0].legend()

# Box plot
axes[1].boxplot(df['G3'], vert=True, patch_artist=True, 
                boxprops=dict(facecolor='lightgreen'))
axes[1].set_title('Box Plot of Final Grades', fontweight='bold')
axes[1].set_ylabel('Grade')

# Performance categories
performance_counts = df['performance_category'].value_counts()
axes[2].pie(performance_counts.values, labels=performance_counts.index, 
            autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99'])
axes[2].set_title('Performance Categories', fontweight='bold')

plt.tight_layout()
plt.savefig('target_variable_analysis.png', dpi=300, bbox_inches='tight')
print("✓ Saved: target_variable_analysis.png")
plt.close()

print(f"\nFinal Grade Statistics:")
print(f"Mean: {df['G3'].mean():.2f}")
print(f"Median: {df['G3'].median():.2f}")
print(f"Std Dev: {df['G3'].std():.2f}")
print(f"Min: {df['G3'].min()}")
print(f"Max: {df['G3'].max()}")

# 4.4 Bivariate Analysis - Numerical vs Target
print("\n--- 4.4 Bivariate Analysis: Numerical Features vs Final Grade ---")

key_numerical = ['studytime', 'failures', 'absences', 'Dalc', 'Walc', 
                 'goout', 'freetime', 'G1', 'G2']

fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.ravel()

for idx, col in enumerate(key_numerical):
    axes[idx].scatter(df[col], df['G3'], alpha=0.5, color='purple')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Final Grade (G3)')
    axes[idx].set_title(f'{col} vs Final Grade')
    
    # Add trend line
    z = np.polyfit(df[col], df['G3'], 1)
    p = np.poly1d(z)
    axes[idx].plot(df[col], p(df[col]), "r--", alpha=0.8, linewidth=2)
    
    # Calculate correlation
    corr = df[col].corr(df['G3'])
    axes[idx].text(0.05, 0.95, f'r = {corr:.3f}', 
                   transform=axes[idx].transAxes, 
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.savefig('bivariate_numerical.png', dpi=300, bbox_inches='tight')
print("✓ Saved: bivariate_numerical.png")
plt.close()

# 4.5 Bivariate Analysis - Categorical vs Target
print("\n--- 4.5 Bivariate Analysis: Categorical Features vs Final Grade ---")

key_categorical = ['school', 'sex', 'address', 'famsize', 'Pstatus', 
                   'schoolsup', 'famsup', 'activities', 'higher']

fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.ravel()

for idx, col in enumerate(key_categorical):
    df.boxplot(column='G3', by=col, ax=axes[idx], patch_artist=True)
    axes[idx].set_title(f'Final Grade by {col}')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Final Grade (G3)')
    axes[idx].get_figure().suptitle('')  # Remove automatic title

plt.tight_layout()
plt.savefig('bivariate_categorical.png', dpi=300, bbox_inches='tight')
print("✓ Saved: bivariate_categorical.png")
plt.close()

# 4.6 Correlation Analysis
print("\n--- 4.6 Correlation Matrix ---")

# Select numerical columns for correlation
corr_cols = numerical_cols + ['total_alcohol', 'study_efficiency', 'family_education']
correlation_matrix = df[corr_cols].corr()

plt.figure(figsize=(16, 14))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Numerical Features', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
print("✓ Saved: correlation_matrix.png")
plt.close()

# Top correlations with G3
print("\nTop 10 Features Correlated with Final Grade (G3):")
g3_correlations = correlation_matrix['G3'].sort_values(ascending=False)
print(g3_correlations.head(11)[1:])  # Exclude G3 itself

# 4.7 Multivariate Analysis
print("\n--- 4.7 Multivariate Analysis ---")

# Pair plot for key features
key_features = ['studytime', 'failures', 'G1', 'G2', 'G3']
pairplot = sns.pairplot(df[key_features], diag_kind='kde', height=2.5)
pairplot.fig.suptitle('Pair Plot of Key Features', y=1.02, fontsize=16, fontweight='bold')
plt.savefig('pairplot.png', dpi=300, bbox_inches='tight')
print("✓ Saved: pairplot.png")
plt.close()

# 4.8 Outlier Detection
print("\n--- 4.8 Outlier Detection ---")

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

outlier_cols = ['age', 'absences', 'G1', 'G2', 'G3', 'failures']

for idx, col in enumerate(outlier_cols):
    axes[idx].boxplot(df[col], vert=True, patch_artist=True,
                      boxprops=dict(facecolor='lightblue'))
    axes[idx].set_title(f'Box Plot: {col}', fontweight='bold')
    axes[idx].set_ylabel(col)
    
    # Calculate and display outliers
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)][col]
    axes[idx].text(0.5, 0.95, f'Outliers: {len(outliers)}', 
                   transform=axes[idx].transAxes, ha='center',
                   bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.5))

plt.tight_layout()
plt.savefig('outlier_detection.png', dpi=300, bbox_inches='tight')
print("✓ Saved: outlier_detection.png")
plt.close()

# ============================================================================
# STAGE 5: FEATURE ENGINEERING
# ============================================================================
print("\n" + "="*80)
print("STAGE 5: FEATURE ENGINEERING")
print("="*80)

# 5.1 Create additional derived features
print("\n--- 5.1 Creating Derived Features ---")

# Family support score (combines parent education and support)
df['family_support_score'] = (df['Medu'] + df['Fedu'] + 
                               df['famsup'].map({'yes': 5, 'no': 0})) / 3

# Social activity score
df['social_score'] = (df['goout'] + df['freetime'] + 
                      df['activities'].map({'yes': 5, 'no': 0})) / 3

# Academic motivation
df['academic_motivation'] = df['higher'].map({'yes': 1, 'no': 0})

# At-risk student (low grades and high absences)
df['at_risk'] = ((df['G3'] < 10) & (df['absences'] > 10)).astype(int)

# Grade improvement
df['grade_improvement'] = df['G3'] - df['G1']

print("✓ Created features: family_support_score, social_score, academic_motivation, at_risk, grade_improvement")

# 5.2 Encoding categorical variables
print("\n--- 5.2 Encoding Categorical Variables ---")

from sklearn.preprocessing import LabelEncoder

categorical_to_encode = ['school', 'sex', 'address', 'famsize', 'Pstatus']

for col in categorical_to_encode:
    le = LabelEncoder()
    df[f'{col}_encoded'] = le.fit_transform(df[col])
    print(f"✓ Encoded {col}: {dict(zip(le.classes_, le.transform(le.classes_)))}")

# 5.3 Feature Importance Analysis
print("\n--- 5.3 Feature Importance (Correlation-based) ---")

# Get all numerical columns including engineered features
all_numerical = df.select_dtypes(include=[np.number]).columns.tolist()
feature_importance = df[all_numerical].corr()['G3'].sort_values(ascending=False)

print("\nTop 15 Most Important Features:")
print(feature_importance.head(16)[1:])  # Exclude G3 itself

# Visualize feature importance
plt.figure(figsize=(10, 12))
feature_importance[1:16].plot(kind='barh', color='teal')
plt.title('Top 15 Features by Correlation with Final Grade', fontsize=14, fontweight='bold')
plt.xlabel('Correlation Coefficient')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300, bbox_inches='tight')
print("\n✓ Saved: feature_importance.png")
plt.close()

# ============================================================================
# STAGE 6: HYPOTHESIS TESTING
# ============================================================================
print("\n" + "="*80)
print("STAGE 6: HYPOTHESIS TESTING")
print("="*80)

results_summary = []

# 6.1 Test 1: Gender and Performance
print("\n--- Test 1: Gender Difference in Performance ---")
print("H0: There is no significant difference in grades between male and female students")
print("H1: There is a significant difference in grades between male and female students")

male_grades = df[df['sex'] == 'M']['G3']
female_grades = df[df['sex'] == 'F']['G3']

t_stat, p_value = ttest_ind(male_grades, female_grades)

print(f"\nMale students - Mean: {male_grades.mean():.2f}, Std: {male_grades.std():.2f}, n: {len(male_grades)}")
print(f"Female students - Mean: {female_grades.mean():.2f}, Std: {female_grades.std():.2f}, n: {len(female_grades)}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    conclusion = "Reject H0: Significant difference exists"
else:
    conclusion = "Fail to reject H0: No significant difference"
print(f"Conclusion (α=0.05): {conclusion}")

results_summary.append({
    'Test': 'Gender vs Performance (T-test)',
    'P-value': p_value,
    'Result': 'Significant' if p_value < alpha else 'Not Significant',
    'Conclusion': conclusion
})

# 6.2 Test 2: Study Time and Grades
print("\n--- Test 2: Study Time vs Performance ---")
print("H0: There is no significant difference in grades across different study time groups")
print("H1: There is a significant difference in grades across different study time groups")

study_groups = [df[df['studytime'] == i]['G3'].values for i in range(1, 5)]
f_stat, p_value = f_oneway(*study_groups)

print(f"\nF-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

for i in range(1, 5):
    group = df[df['studytime'] == i]['G3']
    print(f"Study time {i} - Mean: {group.mean():.2f}, n: {len(group)}")

if p_value < alpha:
    conclusion = "Reject H0: Study time significantly affects grades"
else:
    conclusion = "Fail to reject H0: No significant effect"
print(f"Conclusion (α=0.05): {conclusion}")

results_summary.append({
    'Test': 'Study Time vs Performance (ANOVA)',
    'P-value': p_value,
    'Result': 'Significant' if p_value < alpha else 'Not Significant',
    'Conclusion': conclusion
})

# 6.3 Test 3: Internet Access and Performance
print("\n--- Test 3: Internet Access vs Performance ---")
print("H0: Internet access does not affect student grades")
print("H1: Internet access affects student grades")

internet_yes = df[df['internet'] == 'yes']['G3']
internet_no = df[df['internet'] == 'no']['G3']

t_stat, p_value = ttest_ind(internet_yes, internet_no)

print(f"\nWith Internet - Mean: {internet_yes.mean():.2f}, Std: {internet_yes.std():.2f}, n: {len(internet_yes)}")
print(f"Without Internet - Mean: {internet_no.mean():.2f}, Std: {internet_no.std():.2f}, n: {len(internet_no)}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    conclusion = "Reject H0: Internet access significantly affects grades"
else:
    conclusion = "Fail to reject H0: No significant effect"
print(f"Conclusion (α=0.05): {conclusion}")

results_summary.append({
    'Test': 'Internet Access vs Performance (T-test)',
    'P-value': p_value,
    'Result': 'Significant' if p_value < alpha else 'Not Significant',
    'Conclusion': conclusion
})

# 6.4 Test 4: Alcohol Consumption and Grades
print("\n--- Test 4: Alcohol Consumption vs Performance ---")
print("H0: Alcohol consumption is not correlated with grades")
print("H1: Alcohol consumption is negatively correlated with grades")

corr_coef, p_value = pearsonr(df['total_alcohol'], df['G3'])

print(f"\nPearson Correlation Coefficient: {corr_coef:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    if corr_coef < 0:
        conclusion = "Reject H0: Negative correlation confirmed"
    else:
        conclusion = "Reject H0: Positive correlation found"
else:
    conclusion = "Fail to reject H0: No significant correlation"
print(f"Conclusion (α=0.05): {conclusion}")

results_summary.append({
    'Test': 'Alcohol vs Performance (Pearson)',
    'P-value': p_value,
    'Result': 'Significant' if p_value < alpha else 'Not Significant',
    'Conclusion': conclusion
})

# 6.5 Test 5: Family Education and Performance
print("\n--- Test 5: Family Education Level vs Performance ---")
print("H0: Parents' education level is not correlated with student grades")
print("H1: Parents' education level is positively correlated with student grades")

corr_coef, p_value = pearsonr(df['family_education'], df['G3'])

print(f"\nPearson Correlation Coefficient: {corr_coef:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    if corr_coef > 0:
        conclusion = "Reject H0: Positive correlation confirmed"
    else:
        conclusion = "Reject H0: Negative correlation found"
else:
    conclusion = "Fail to reject H0: No significant correlation"
print(f"Conclusion (α=0.05): {conclusion}")

results_summary.append({
    'Test': 'Family Education vs Performance (Pearson)',
    'P-value': p_value,
    'Result': 'Significant' if p_value < alpha else 'Not Significant',
    'Conclusion': conclusion
})

# 6.6 Test 6: Failures and Current Performance
print("\n--- Test 6: Past Failures vs Current Performance ---")
print("H0: There is no significant difference in grades based on past failures")
print("H1: There is a significant difference in grades based on past failures")

no_failures = df[df['failures'] == 0]['G3']
with_failures = df[df['failures'] > 0]['G3']

t_stat, p_value = ttest_ind(no_failures, with_failures)

print(f"\nNo failures - Mean: {no_failures.mean():.2f}, Std: {no_failures.std():.2f}, n: {len(no_failures)}")
print(f"With failures - Mean: {with_failures.mean():.2f}, Std: {with_failures.std():.2f}, n: {len(with_failures)}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    conclusion = "Reject H0: Past failures significantly affect current grades"
else:
    conclusion = "Fail to reject H0: No significant effect"
print(f"Conclusion (α=0.05): {conclusion}")

results_summary.append({
    'Test': 'Past Failures vs Performance (T-test)',
    'P-value': p_value,
    'Result': 'Significant' if p_value < alpha else 'Not Significant',
    'Conclusion': conclusion
})

# 6.7 Test 7: School Support and Performance
print("\n--- Test 7: Extra Educational Support vs Performance ---")
print("H0: Extra educational support does not affect student grades")
print("H1: Extra educational support affects student grades")

support_yes = df[df['schoolsup'] == 'yes']['G3']
support_no = df[df['schoolsup'] == 'no']['G3']

t_stat, p_value = ttest_ind(support_yes, support_no)

print(f"\nWith support - Mean: {support_yes.mean():.2f}, Std: {support_yes.std():.2f}, n: {len(support_yes)}")
print(f"Without support - Mean: {support_no.mean():.2f}, Std: {support_no.std():.2f}, n: {len(support_no)}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    conclusion = "Reject H0: Educational support significantly affects grades"
else:
    conclusion = "Fail to reject H0: No significant effect"
print(f"Conclusion (α=0.05): {conclusion}")

results_summary.append({
    'Test': 'School Support vs Performance (T-test)',
    'P-value': p_value,
    'Result': 'Significant' if p_value < alpha else 'Not Significant',
    'Conclusion': conclusion
})

# Summary of All Tests
print("\n" + "="*80)
print("HYPOTHESIS TESTING SUMMARY")
print("="*80)

summary_df = pd.DataFrame(results_summary)
print("\n", summary_df.to_string(index=False))

# Visualize hypothesis testing results
fig, ax = plt.subplots(figsize=(12, 6))
colors = ['green' if r == 'Significant' else 'red' for r in summary_df['Result']]
bars = ax.barh(summary_df['Test'], -np.log10(summary_df['P-value']), color=colors, alpha=0.7)
ax.axvline(x=-np.log10(0.05), color='black', linestyle='--', linewidth=2, label='α = 0.05')
ax.set_xlabel('-log10(P-value)', fontsize=12, fontweight='bold')
ax.set_title('Hypothesis Testing Results', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# Add p-value annotations
for i, (test, pval) in enumerate(zip(summary_df['Test'], summary_df['P-value'])):
    ax.text(-np.log10(pval) + 0.1, i, f'p={pval:.4f}', va='center')

plt.tight_layout()
plt.savefig('hypothesis_testing_summary.png', dpi=300, bbox_inches='tight')
print("\n✓ Saved: hypothesis_testing_summary.png")
plt.close()

# ============================================================================
# STAGE 7: INSIGHTS & RECOMMENDATIONS
# ============================================================================
print("\n" + "="*80)
print("STAGE 7: INSIGHTS & RECOMMENDATIONS")
print("="*80)

print("\n📊 KEY INSIGHTS:")
print("-" * 80)

# Insight 1: Grade progression
print("\n1. GRADE PROGRESSION:")
g1_g3_corr = df['G1'].corr(df['G3'])
print(f"   - Strong correlation between G1 and G3: r = {g1_g3_corr:.3f}")
print(f"   - Early intervention is crucial for struggling students")

# Insight 2: Study time impact
study_effect = df.groupby('studytime')['G3'].mean()
print(f"\n2. STUDY TIME IMPACT:")
print(f"   - Average grades by study time:")
for st, grade in study_effect.items():
    print(f"     Study time {st}: {grade:.2f}")

# Insight 3: Failure impact
failure_impact = df.groupby('failures')['G3'].mean()
print(f"\n3. PAST FAILURES:")
print(f"   - Students with no failures: {failure_impact[0]:.2f} average")
print(f"   - Students with failures: {failure_impact[failure_impact.index > 0].mean():.2f} average")

# Insight 4: Alcohol consumption
high_alcohol = df[df['total_alcohol'] > 6]['G3'].mean()
low_alcohol = df[df['total_alcohol'] <= 3]['G3'].mean()
print(f"\n4. ALCOHOL CONSUMPTION:")
print(f"   - High consumption (>6): {high_alcohol:.2f} average grade")
print(f"   - Low consumption (≤3): {low_alcohol:.2f} average grade")
print(f"   - Difference: {low_alcohol - high_alcohol:.2f} points")

# Insight 5: Internet access
internet_impact = df.groupby('internet')['G3'].mean()
print(f"\n5. INTERNET ACCESS:")
for status, grade in internet_impact.items():
    print(f"   - {status.capitalize()}: {grade:.2f} average grade")

# Insight 6: Family education
high_fam_edu = df[df['family_education'] >= 3]['G3'].mean()
low_fam_edu = df[df['family_education'] < 3]['G3'].mean()
print(f"\n6. FAMILY EDUCATION LEVEL:")
print(f"   - High family education (≥3): {high_fam_edu:.2f} average")
print(f"   - Low family education (<3): {low_fam_edu:.2f} average")

# Insight 7: At-risk students
at_risk_count = df['at_risk'].sum()
at_risk_pct = (at_risk_count / len(df)) * 100
print(f"\n7. AT-RISK STUDENTS:")
print(f"   - Number of at-risk students: {at_risk_count} ({at_risk_pct:.1f}%)")
print(f"   - Definition: Grade < 10 AND Absences > 10")

print("\n\n💡 RECOMMENDATIONS:")
print("-" * 80)

recommendations = [
    {
        'area': 'EARLY INTERVENTION',
        'recommendation': 'Implement monitoring system for G1 grades',
        'rationale': f'Strong correlation (r={g1_g3_corr:.3f}) between first and final grades',
        'action': 'Identify struggling students early and provide additional support'
    },
    {
        'area': 'STUDY HABITS',
        'recommendation': 'Promote structured study schedules',
        'rationale': 'Higher study time correlates with better performance',
        'action': 'Create study groups and time management workshops'
    },
    {
        'area': 'FAILURE PREVENTION',
        'recommendation': 'Develop remedial programs for students with past failures',
        'rationale': f'Students with failures score {failure_impact[0] - failure_impact[failure_impact.index > 0].mean():.2f} points lower',
        'action': 'Personalized learning plans and tutoring services'
    },
    {
        'area': 'HEALTH & WELLNESS',
        'recommendation': 'Implement awareness programs on alcohol effects',
        'rationale': f'High alcohol consumption associated with {low_alcohol - high_alcohol:.2f} point decrease',
        'action': 'Counseling services and health education campaigns'
    },
    {
        'area': 'DIGITAL RESOURCES',
        'recommendation': 'Ensure internet access for all students',
        'rationale': f'Students with internet score {internet_impact["yes"] - internet_impact["no"]:.2f} points higher on average',
        'action': 'Partner with libraries, provide computer labs, or subsidize internet'
    },
    {
        'area': 'FAMILY ENGAGEMENT',
        'recommendation': 'Create parent education programs',
        'rationale': f'Higher family education linked to {high_fam_edu - low_fam_edu:.2f} point advantage',
        'action': 'Parent workshops on supporting student learning'
    },
    {
        'area': 'AT-RISK IDENTIFICATION',
        'recommendation': 'Deploy early warning system',
        'rationale': f'{at_risk_pct:.1f}% of students identified as at-risk',
        'action': 'Regular monitoring of grades and attendance with intervention protocols'
    }
]

for i, rec in enumerate(recommendations, 1):
    print(f"\n{i}. {rec['area']}")
    print(f"   Recommendation: {rec['recommendation']}")
    print(f"   Rationale: {rec['rationale']}")
    print(f"   Action: {rec['action']}")

# Create comprehensive visualization dashboard
print("\n\n📈 Creating Comprehensive Dashboard...")

fig = plt.figure(figsize=(20, 12))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. Grade Distribution
ax1 = fig.add_subplot(gs[0, 0])
ax1.hist(df['G3'], bins=20, edgecolor='black', alpha=0.7, color='#2ecc71')
ax1.set_title('Final Grade Distribution', fontweight='bold', fontsize=12)
ax1.set_xlabel('Grade')
ax1.set_ylabel('Frequency')
ax1.axvline(df['G3'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {df["G3"].mean():.2f}')
ax1.legend()

# 2. Study Time Effect
ax2 = fig.add_subplot(gs[0, 1])
study_effect.plot(kind='bar', ax=ax2, color='#3498db', edgecolor='black')
ax2.set_title('Study Time vs Average Grade', fontweight='bold', fontsize=12)
ax2.set_xlabel('Study Time Level')
ax2.set_ylabel('Average Grade')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=0)
ax2.grid(True, alpha=0.3, axis='y')

# 3. Gender Comparison
ax3 = fig.add_subplot(gs[0, 2])
df.boxplot(column='G3', by='sex', ax=ax3, patch_artist=True)
ax3.set_title('Grades by Gender', fontweight='bold', fontsize=12)
ax3.set_xlabel('Gender')
ax3.set_ylabel('Final Grade')
ax3.get_figure().suptitle('')

# 4. Alcohol Impact
ax4 = fig.add_subplot(gs[1, 0])
alcohol_bins = pd.cut(df['total_alcohol'], bins=[0, 3, 6, 10], labels=['Low', 'Medium', 'High'])
df.groupby(alcohol_bins)['G3'].mean().plot(kind='bar', ax=ax4, color='#e74c3c', edgecolor='black')
ax4.set_title('Alcohol Consumption vs Grade', fontweight='bold', fontsize=12)
ax4.set_xlabel('Alcohol Consumption Level')
ax4.set_ylabel('Average Grade')
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=0)
ax4.grid(True, alpha=0.3, axis='y')

# 5. Failures Impact
ax5 = fig.add_subplot(gs[1, 1])
failures_grouped = df.groupby('failures')['G3'].mean()
failures_grouped.plot(kind='bar', ax=ax5, color='#9b59b6', edgecolor='black')
ax5.set_title('Past Failures vs Current Grade', fontweight='bold', fontsize=12)
ax5.set_xlabel('Number of Past Failures')
ax5.set_ylabel('Average Grade')
ax5.set_xticklabels(ax5.get_xticklabels(), rotation=0)
ax5.grid(True, alpha=0.3, axis='y')

# 6. Internet Access
ax6 = fig.add_subplot(gs[1, 2])
internet_grouped = df.groupby('internet')['G3'].mean()
internet_grouped.plot(kind='bar', ax=ax6, color='#1abc9c', edgecolor='black')
ax6.set_title('Internet Access vs Grade', fontweight='bold', fontsize=12)
ax6.set_xlabel('Internet Access')
ax6.set_ylabel('Average Grade')
ax6.set_xticklabels(ax6.get_xticklabels(), rotation=0)
ax6.grid(True, alpha=0.3, axis='y')

# 7. Family Education
ax7 = fig.add_subplot(gs[2, 0])
ax7.scatter(df['family_education'], df['G3'], alpha=0.5, color='#f39c12')
z = np.polyfit(df['family_education'], df['G3'], 1)
p = np.poly1d(z)
ax7.plot(df['family_education'], p(df['family_education']), "r-", linewidth=2)
ax7.set_title('Family Education vs Grade', fontweight='bold', fontsize=12)
ax7.set_xlabel('Family Education Level')
ax7.set_ylabel('Final Grade')
ax7.grid(True, alpha=0.3)

# 8. Grade Progression
ax8 = fig.add_subplot(gs[2, 1])
grade_progression = df[['G1', 'G2', 'G3']].mean()
grade_progression.plot(kind='line', ax=ax8, marker='o', linewidth=3, markersize=10, color='#e67e22')
ax8.set_title('Average Grade Progression', fontweight='bold', fontsize=12)
ax8.set_xlabel('Period')
ax8.set_ylabel('Average Grade')
ax8.set_xticklabels(['G1', 'G2', 'G3'])
ax8.grid(True, alpha=0.3)

# 9. Performance Categories
ax9 = fig.add_subplot(gs[2, 2])
performance_counts = df['performance_category'].value_counts()
colors_pie = ['#e74c3c', '#f39c12', '#2ecc71']
ax9.pie(performance_counts.values, labels=performance_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=colors_pie)
ax9.set_title('Performance Categories', fontweight='bold', fontsize=12)

fig.suptitle('Student Performance Analysis Dashboard', fontsize=20, fontweight='bold', y=0.98)
plt.savefig('comprehensive_dashboard.png', dpi=300, bbox_inches='tight')
print("✓ Saved: comprehensive_dashboard.png")
plt.close()

# ============================================================================
# FINAL REPORT GENERATION
# ============================================================================
print("\n" + "="*80)
print("GENERATING FINAL REPORT")
print("="*80)

report = f"""
{'='*80}
STUDENT PERFORMANCE ANALYSIS - FINAL REPORT
Dataset: Student Performance in Portuguese Language Course
{'='*80}

EXECUTIVE SUMMARY
{'-'*80}
This analysis examined {len(df)} student records with {df.shape[1]} features to identify
factors affecting academic performance in Portuguese language course.

KEY STATISTICS:
- Average Final Grade (G3): {df['G3'].mean():.2f} ± {df['G3'].std():.2f}
- Grade Range: {df['G3'].min()} to {df['G3'].max()}
- Students at risk (Grade < 10, Absences > 10): {at_risk_count} ({at_risk_pct:.1f}%)
- Performance Distribution:
  * Poor (0-9): {len(df[df['performance_category']=='Poor'])} students ({len(df[df['performance_category']=='Poor'])/len(df)*100:.1f}%)
  * Average (10-14): {len(df[df['performance_category']=='Average'])} students ({len(df[df['performance_category']=='Average'])/len(df)*100:.1f}%)
  * Good (15-20): {len(df[df['performance_category']=='Good'])} students ({len(df[df['performance_category']=='Good'])/len(df)*100:.1f}%)

DATA QUALITY ASSESSMENT
{'-'*80}
✓ No missing values detected
✓ No duplicate records found
✓ {len(df.columns)} features analyzed
✓ {len(numerical_cols)} numerical features
✓ {len(categorical_cols)} categorical features

TOP FACTORS AFFECTING PERFORMANCE (by correlation)
{'-'*80}
{g3_correlations.head(11)[1:].to_string()}

HYPOTHESIS TESTING RESULTS
{'-'*80}
Conducted 7 statistical tests at α = 0.05 significance level:

"""

for i, result in enumerate(results_summary, 1):
    report += f"{i}. {result['Test']}\n"
    report += f"   P-value: {result['P-value']:.4f} - {result['Result']}\n"
    report += f"   {result['Conclusion']}\n\n"

report += f"""
CRITICAL FINDINGS
{'-'*80}
1. EARLY PREDICTION POWER
   - G1 (first period grade) shows correlation of {df['G1'].corr(df['G3']):.3f} with final grade
   - Early intervention based on G1 scores can prevent failure

2. STUDY TIME IMPACT
   - Students with highest study time (level 4) average: {study_effect[4]:.2f}
   - Students with lowest study time (level 1) average: {study_effect[1]:.2f}
   - Difference: {study_effect[4] - study_effect[1]:.2f} points

3. FAILURE HISTORY
   - Students with no failures: {failure_impact[0]:.2f} average grade
   - Students with 1+ failures: {failure_impact[failure_impact.index > 0].mean():.2f} average grade
   - Impact: {failure_impact[0] - failure_impact[failure_impact.index > 0].mean():.2f} points lower

4. ALCOHOL CONSUMPTION
   - Low consumption (≤3): {low_alcohol:.2f} average
   - High consumption (>6): {high_alcohol:.2f} average
   - Impact: {low_alcohol - high_alcohol:.2f} points difference

5. INTERNET ACCESS
   - With internet: {internet_impact['yes']:.2f} average
   - Without internet: {internet_impact['no']:.2f} average
   - Digital divide impact: {internet_impact['yes'] - internet_impact['no']:.2f} points

6. FAMILY EDUCATION
   - High family education (≥3): {high_fam_edu:.2f} average
   - Low family education (<3): {low_fam_edu:.2f} average
   - Socioeconomic impact: {high_fam_edu - low_fam_edu:.2f} points

ACTIONABLE RECOMMENDATIONS
{'-'*80}
"""

for i, rec in enumerate(recommendations, 1):
    report += f"{i}. {rec['area']}\n"
    report += f"   → {rec['recommendation']}\n"
    report += f"   Rationale: {rec['rationale']}\n"
    report += f"   Action: {rec['action']}\n\n"

report += f"""
VISUALIZATIONS GENERATED
{'-'*80}
✓ univariate_numerical.png - Distribution of numerical features
✓ univariate_categorical.png - Distribution of categorical features
✓ target_variable_analysis.png - Final grade distribution analysis
✓ bivariate_numerical.png - Numerical features vs final grade
✓ bivariate_categorical.png - Categorical features vs final grade
✓ correlation_matrix.png - Feature correlation heatmap
✓ pairplot.png - Pairwise relationships of key features
✓ outlier_detection.png - Box plots for outlier identification
✓ feature_importance.png - Top features by correlation
✓ hypothesis_testing_summary.png - Statistical test results
✓ comprehensive_dashboard.png - Executive dashboard

DATA SCIENCE PROCESS COMPLETION
{'-'*80}
✓ Stage 1: Problem Definition - Completed
✓ Stage 2: Data Collection - Completed
✓ Stage 3: Data Preparation & Cleaning - Completed
✓ Stage 4: Exploratory Data Analysis - Completed
✓ Stage 5: Feature Engineering - Completed
✓ Stage 6: Hypothesis Testing - Completed
✓ Stage 7: Insights & Recommendations - Completed

{'='*80}
Report Generated: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}
{'='*80}
"""

# Save report to file
with open('final_report.txt', 'w') as f:
    f.write(report)

print(report)
print("\n✓ Saved: final_report.txt")

# Save processed dataset
df.to_csv('student_performance_processed.csv', index=False)
print("✓ Saved: student_performance_processed.csv (with engineered features)")

# Create summary statistics CSV
summary_stats = pd.DataFrame({
    'Metric': ['Total Students', 'Average Final Grade', 'Median Final Grade', 
               'Std Dev Final Grade', 'Min Grade', 'Max Grade',
               'At-Risk Students', 'At-Risk Percentage'],
    'Value': [len(df), df['G3'].mean(), df['G3'].median(), 
              df['G3'].std(), df['G3'].min(), df['G3'].max(),
              at_risk_count, f"{at_risk_pct:.1f}%"]
})
summary_stats.to_csv('summary_statistics.csv', index=False)
print("✓ Saved: summary_statistics.csv")

# Save hypothesis testing results
summary_df.to_csv('hypothesis_testing_results.csv', index=False)
print("✓ Saved: hypothesis_testing_results.csv")

print("\n" + "="*80)
print("ANALYSIS COMPLETE! 🎉")
print("="*80)
print("\nGenerated Files:")
print("  1. final_report.txt - Comprehensive analysis report")
print("  2. student_performance_processed.csv - Dataset with engineered features")
print("  3. summary_statistics.csv - Key metrics summary")
print("  4. hypothesis_testing_results.csv - Statistical test results")
print("  5-15. Various PNG visualization files")
print("\nAll 7 stages of the Data Science process completed successfully!")
print("="*80)
```
